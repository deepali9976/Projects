```python
class Node:
    def __init__(self, count, code):
        self.count = count
        self.next = None
        self.code = code

class LinkedList:
    def __init__(self):
        self.head = None

    def CountCheck(self):
        count = 0
        current = self.head
        while current is not None:
            count += 1
            current = current.next
        return count

    def AddItem(self, count, code):
        new = Node(count, code)
        if self.head is None:
            self.head = new
        else:
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new

    def FindItem(self, code):
        current = self.head
        while current is not None:
            if current.code == code:
                print("Item found")
                return
            current = current.next
        print("Item not found")
    def RemoveItem(self, code):
      if self.head is None:
          print("No items in the list")
          return

      if self.head.code == code:
          self.head = self.head.next
          print("Item successfully removed")
          return

      current = self.head
      while current.next is not None and current.next.code != code:
          current = current.next

      if current.next is None:
          print("Item not found")
      else:
        current.next = current.next.next
        print("Item successfully removed")




trousers = LinkedList()
shoes = LinkedList()


sweaters = LinkedList()
sweaters.AddItem(5, "123")
sweaters.AddItem(3, "456")
trousers = LinkedList()
trousers.AddItem(6, "12345")
trousers.AddItem(7, "456789")
shoes = LinkedList()
shoes.AddItem(1, "123123")
shoes.AddItem(2, "456456")


n = int(input("Select operation\n1. Check quantity of required item\n2. Add item\n3. Find item\n4.Remove Item\n5.Total number of items in the inventory\n"))

if n<5:
  i = int(input("Select item\n1. Sweaters\n2. Trousers\n3. Shoes\n"))

  if n == 1:
      if i == 1:
          disp = sweaters.CountCheck()
          print(f"There are {disp} sweaters in the store")
      elif i == 2:
          disp = trousers.CountCheck()
          print(f"There are {disp} trousers in the store")
      elif i == 3:
          disp = shoes.CountCheck()
          print(f"There are {disp} shoes in the store")
      else:
          print("No such item found")
  elif n == 2:
      count = int(input("Enter count of the item to be added: "))
      code = input("Enter code of the item to be added: ")
      if i == 1:
         sweaters.AddItem(count, code)
      elif i == 2:
        trousers.AddItem(count, code)
      elif i == 3:
        shoes.AddItem(count, code)
      else:
         print("Error! Enter valid item")
      print("Item Successfully Added!")
  elif n == 3:
    if i<=3:
       code = input("Enter code of the item to find: ")
       if i==1:
             sweaters.FindItem(code)
       elif i == 2:
            trousers.FindItem(code)
       elif i == 3:
            shoes.FindItem(code)
    else:
       print("Error! Enter valid item")
  elif n==4:
    if i<3:

       code = input("Enter code of the item to be removed: ")
       if i==1:
             sweaters.RemoveItem(code)
       elif i == 2:
            trousers.RemoveItem(code)
       elif i == 3:
            shoes.RemoveItem(code)
    else:
       print("Error! Enter valid item")
elif n==5:
   sw=sweaters.CountCheck()
   tr=trousers.CountCheck()
   sh= shoes.CountCheck()
   n=sw+tr+sh
   print(f"There are {n} items in stock in the inventory")
else:
    print("Invalid operation")
```
